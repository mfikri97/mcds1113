```python
#General Dictonary maker
def category_dict_maker(js):
    items = js['items']
    item_id = []
    item_snippet_title = []
    for item in items:
        item_id.append(item['id']) 
        item_snippet_title.append(str(item['snippet']['title']))
    item_dict = dict(zip(item_id, item_snippet_title))
    return(item_dict)
```

```python
#Loading json file to map categories
with open('CA_category_id.json') as f:
    canada_js = json.load(f)
with open('DE_category_id.json') as f:
    denmark_js = json.load(f)
with open('FR_category_id.json') as f:
    france_js = json.load(f)
with open('GB_category_id.json') as f:
    britain_js = json.load(f)
with open('IN_category_id.json') as f:
    india_js = json.load(f)
with open('JP_category_id.json') as f:
    japan_js = json.load(f)
with open('KR_category_id.json') as f:
    korea_js = json.load(f)
with open('MX_category_id.json') as f:
    mexico_js = json.load(f)
with open('RU_category_id.json') as f:
    russia_js = json.load(f)
with open('US_category_id.json') as f:
    us_js = json.load(f)
```

```python
#Assigning dictionary to each country
can_dict = category_dict_maker(canada_js)
den_dict = category_dict_maker(denmark_js)
fra_dict = category_dict_maker(france_js)
bri_dict = category_dict_maker(britain_js)
ind_dict = category_dict_maker(india_js)
jap_dict = category_dict_maker(japan_js)
kor_dict = category_dict_maker(korea_js)
mex_dict = category_dict_maker(mexico_js)
rus_dict = category_dict_maker(russia_js)
usa_dict = category_dict_maker(us_js)
```

```python
#Category Maker function
def category_maker(value):
    for key in can_dict:
        if str(value) == key:
            return (can_dict[key])
        else:
            continue
            
    for key in den_dict:
        if str(value) == key:
            return (den_dict[key])
        else:
            continue
            
    for key in fra_dict:
        if str(value) == key:
            return (fra_dict[key])
        else:
            continue
    for key in bri_dict:
        if str(value) == key:
            return (bri_dict[key])
        else:
            continue
    for key in ind_dict:
        if str(value) == key:
            return (ind_dict[key])
        else:
            continue
    for key in jap_dict:
        if str(value) == key:
            return (jap_dict[key])
        else:
            continue
    for key in kor_dict:
        if str(value) == key:
            return (kor_dict[key])
        else:
            continue
    for key in mex_dict:
        if str(value) == key:
            return (mex_dict[key])
        else:
            continue
    for key in rus_dict:
        if str(value) == key:
            return (rus_dict[key])
        else:
            continue
    for key in usa_dict:
        if str(value) == key:
            return (usa_dict[key])
        else:
            continue
        
    
```

```python
#Category Function while dropping category id column
def cat_applier(df):
    df['category'] = df.category_id.apply(func=category_maker)
    df.category = df.category.astype('category')
    return df.drop(columns=['category_id'])
```

```python
#Fixing Data Types
for df in all_df:
    # video_id 
    df['video_id'] = df['video_id'].astype('str') 
    
    #title
    df['title'] = df['title'].astype('str')
    
    #channel_title
    df['channel_title'] = df['channel_title'].astype('str')
    
    #tags
    df['tags'] = df['tags'].astype('str')
    
    # views, likes, dislikes, comment_count are already in correct data types i.e int64
    
    #thumbnail_link 
    df['thumbnail_link'] = df['thumbnail_link'].astype('str') 
    
    #description
    df['description'] = df['description'].astype('str')
    
    # Changing comments_disabled, ratings_disabled, video_error_or_removed from bool to categorical
    df['comments_disabled'] = df['comments_disabled'].astype('category') 
    df['ratings_disabled'] = df['ratings_disabled'].astype('category') 
    df['video_error_or_removed'] = df['video_error_or_removed'].astype('category') 
    
    # publish_time 
    df['publish_time'] = pd.to_datetime(df['publish_time'], errors='coerce', format='%Y-%m-%dT%H:%M:%S.%fZ')
    df.insert(4, 'publish_date', df['publish_time'].dt.date)
    # loc, column name, values for column to be inserted
    df['publish_time'] = df['publish_time'].dt.time

    #trending date
    df['trending_date'] = df['trending_date'].astype('str') 
    date_pieces = (df['trending_date'].str.split('.'))
    df['Year'] = date_pieces.str[0].astype(int)
    df['Day'] = date_pieces.str[1].astype(int)
    df['Month'] = date_pieces.str[2].astype(int)
    updatedyear = []
    for i in range(len(df)) : 
        y = df.loc[i, "Year"]
        newy = y+2000
        updatedyear.append(newy)
    for i in range(len(df)):
        newy = updatedyear[i]
        tr = df.loc[i, "Year"]
        df['Year'].replace(to_replace = tr, value = newy, inplace=True)
    del df['trending_date']
    df['trending_date'] = pd.to_datetime(df[['Year', 'Month', 'Day']], format = "%Y-%m-%d")
    del df['Year']
    del df['Day']
    del df['Month']
    
    #Dropping Unwanted columns
    df.drop(columns=['description', 'ratings_disabled', 'thumbnail_link'])
```

<font size=5><b>1. Ratio of likes-dislikes in different categories</b></font>

```python
like_dislike_ratio = finalcombined_df.drop_duplicates(subset='video_id',keep='first',inplace=False)\
.groupby(['category'], as_index=False).agg({'likes': 'sum', 'dislikes': 'sum'})
like_dislike_ratio['total'] = like_dislike_ratio.apply(lambda data: data.likes + data.dislikes , axis=1)
like_dislike_ratio['likes_ratio'] = like_dislike_ratio.apply(lambda data: data.likes * 100 / data.total, axis=1)
like_dislike_ratio['dislikes_ratio'] = like_dislike_ratio.apply(lambda data: data.dislikes * 100 / data.total, axis=1)
like_dislike_ratio.sort_values(['likes_ratio'], ascending=False)
```
